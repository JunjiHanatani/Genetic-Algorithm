```python
df_energy = pd.read_csv("./breathing/energy.csv")
sns.set_palette("Set1", 4)
import matplotlib
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['axes.titlesize'] = 18
matplotlib.rcParams['axes.labelsize'] = 18
matplotlib.rcParams['xtick.labelsize'] = 14
matplotlib.rcParams['ytick.labelsize'] = 14
matplotlib.rcParams['legend.loc'] = 'best'
matplotlib.rcParams['legend.frameon'] = True
matplotlib.rcParams['legend.fontsize'] = 12
matplotlib.rcParams['legend.edgecolor'] = 'k'

labels = ["Elastic Potential Energy", 
          "Gravitational Potential Energy", 
          "Contact Potential Energy",
          "Kinetic Energy"]

fig = plt.figure(figsize=[8, 6])
ax = fig.add_subplot(111)
ax.stackplot(df_energy["time"],
             df_energy["Elastic Potential Energy"],
             df_energy["Gravitational Potential Energy"],
             df_energy["Contact Potential Energy"],
             df_energy["Kinetic Energy"],
             alpha = 0.3,
             labels=labels)
ax.plot(df_energy["time"],
        df_energy["Elastic Potential Energy"] + df_energy["External Energy"])
ax.plot(df_energy["time"],
        df_energy["Elastic Potential Energy"]+df_energy["Gravitational Potential Energy"])
ax.plot(df_energy["time"],
        df_energy["Elastic Potential Energy"]
        +df_energy["Gravitational Potential Energy"]
        +df_energy["Contact Potential Energy"])
ax.plot(df_energy["time"],
        df_energy["Total Energy"])

#ax.set_xlim([0.0, 0.16])
ax.set_ylim([0.0, 1.5])
ax.set_title("Energy plot (Breathing Cube)")
ax.set_xlabel("Time[sec.]")
ax.set_ylabel("Energy[J]")
ax.legend(loc="best")
plt.show()

err_rate = (df_energy["Total Energy"].max() - df_energy["Total Energy"].mean())/df_energy["Total Energy"].mean() * 100
print("Err = {:4.2f}%".format(err_rate))
```
